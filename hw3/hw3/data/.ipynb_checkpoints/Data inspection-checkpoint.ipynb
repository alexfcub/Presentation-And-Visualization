```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("WA_Retail-SalesMarketing_-ProfitCost.csv")
#X = df_see.drop("gene",axis=1).values.astype("int") # to drop the gene column and convert data to int
samples = df.columns
df.head()
```

```python
year_profit = df
year_profit = year_profit[['Year', 'Gross profit']].groupby(['Year']).sum()
year_profit['Gross profit'] = np.round(year_profit['Gross profit']/1000000, 2)
year_profit
```

```python
year_profit.to_csv('data/year_profit.csv')
```

```python
year_profit = df[df['Year']==2007]
year_profit = year_profit[['Year','Product line', 'Gross profit']].groupby(['Year', 'Product line']).sum()
year_profit['Gross profit'] = np.round(year_profit['Gross profit']/1000000, 2)
year_profit
```

```python
year_profit.to_csv('data/year_product_profit.csv')
```

```python
df_2007 = df[df["Year"]==2007]
df_2007
```

```python
df_2007 = df_2007.fillna(0)
df_2007
```

```python
gross_by_product_line = df_2007.groupby("Product line",as_index = False).sum()[["Product line","Revenue","Gross profit"]]
gross_by_product_line
```

```python
gross_by_product_line.to_csv("data/gross_by_product_line.csv")
```

```python
gross_by_country = df_2007.groupby("Retailer country",as_index = False).sum()[["Retailer country","Revenue","Gross profit"]]
gross_by_country['Gross profit'] = np.round(gross_by_country['Gross profit']/1000000,2)
gross_by_country = gross_by_country.sort_values(['Gross profit'], ascending=False)
gross_by_country
```

```python
gross_by_country.to_csv("data/gross_by_country.csv", index=False)
```

```python
country_product_gross_6= df_2007.groupby(["Retailer country","Product line"],as_index = False).sum()[["Retailer country","Product line","Gross profit"]]
country_product_gross_6['Gross profit'] = np.round(country_product_gross_6['Gross profit']/1000000,2)
country_product_gross_6
```

```python
table = pd.pivot_table(country_product_gross_6, values='Gross profit', index=['Retailer country'],columns=['Product line'], aggfunc=np.sum)
table
```

```python
table.to_csv("data/country_product_gross_6.csv")
```

```python
df_2007 = df[df["Year"]==2007]
product_line = df['Product line'].unique()
max_product_line = []
for pl in product_line:
    df_2007 = df[df['Product line'] == pl]
    df_2007 = df_2007[['Product line', 'Product', 'Gross profit']]
    df_2007 = df_2007.fillna(0)
    df_2007 = df_2007.groupby(['Product line', 'Product']).sum()
    
    max_product_line.append(df_2007['Gross profit'].argmax())
max_product_line
df_max_line = pd.DataFrame.from_records(max_product_line, columns=['Product line', 'Product'])
df_max_line
df_max_line.to_csv('data/max_product_line.csv')
```

```python
df_2007 = df[df["Year"]==2007]
product_line = df['Retailer country'].unique()
max_product_line = []
for pl in product_line:
    df_2007 = df[df['Retailer country'] == pl]
    df_2007 = df_2007[['Retailer country', 'Product', 'Gross profit']]
    df_2007 = df_2007.fillna(0)
    df_2007 = df_2007.groupby(['Retailer country', 'Product']).sum()
    
    max_product_line.append(df_2007['Gross profit'].argmax())
max_product_line
df_max_country = pd.DataFrame.from_records(max_product_line, columns=['Retailer country', 'Product'])
df_max_country
df_max_country.to_csv('data/max_country.csv')
```
