### country - money  (each year) 

```python
C = df[["Year","Retailer country","Revenue","Planned revenue","Gross profit"]]
```

```python
C7 = C[C["Year"]==2007]
C7 = C7.drop(["Year"],axis=1)
C7 = C7.groupby("Retailer country").sum()
C7 = C7.fillna(0)
C7.to_csv("dataM/country2007.csv")
```

```python
C4 = C[C["Year"]==2004]
C4 = C4.drop(["Year"],axis=1)
C4 = C4.groupby("Retailer country").sum()
C4 = C4.fillna(0)
C4.to_csv("dataM/country2004.csv")
```

```python
C5 = C[C["Year"]==2005]
C5 = C5.drop(["Year"],axis=1)
C5 = C5.groupby("Retailer country").sum()
C5 = C5.fillna(0)
C5.to_csv("dataM/country2005.csv")
```

```python
C6 = C[C["Year"]==2006]
C6 = C6.drop(["Year"],axis=1)
C6 = C6.groupby("Retailer country").sum()
C6 = C6.fillna(0)
C6.to_csv("dataM/country2006.csv")
```
