```python
df_results = pd.read_csv("./2016-us-election/primary_results.csv")
```

```python
df_results[df_results["state_abbreviation"]=="AL"]
```

```python
candidates = df_results['candidate'].unique()
candidates
```

```python
county_id = 'fips'
df_ = df_results[[county_id,'votes','candidate']]
df_pivot = df_.pivot_table(index=county_id, columns='candidate', values='votes').fillna(0.0).astype('int')
df_pivot.head()
```

```python
df_pivot['votes']=df_pivot.sum(axis = 1)
df_pivot[county_id] = df_pivot.index
for candidate in candidates:
    df_pivot[candidate + ' prop'] = df_pivot[candidate]/df_pivot['votes']
df_pivot.to_csv('county_candidate_votes.csv')
df_pivot.head()
```

```python
df_facts.head()
```

```python
df = pd.merge(df_facts, df_pivot, how = 'left', left_on=county_id, right_on=county_id)
df.head()
```

```python
df.to_csv('county_candidate_votes_and_facts.csv')
```
